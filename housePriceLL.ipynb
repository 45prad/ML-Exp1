```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/housing.csv');
```

dropping the rows with null values


```python
df.dropna(inplace=True);
```

One-hot-encoder: The features are encoded using a one-hot (aka ‘one-of-K’ or ‘dummy’) encoding scheme. This creates a binary column for
each category and returns a sparse matrix or dense array


```python
pd.get_dummies(df.ocean_proximity)
```

```python
df = df.join(pd.get_dummies(df.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

**Exploratory Data Analysis**

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm') #Correlation heatmap
plt.show()
```

 **Model Development**

```python
x = df.drop(['median_house_value'], axis=1)
y = df['median_house_value']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Scaling the values

```python
# Feature scaling using standard scaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred = lr.predict(x_test)
print(type(y_pred))
print(type(y_test))
```

```python
y_test = y_test.to_numpy()
```

```python
print(type(y_test))
```

**Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {r2}')
```

```python
#CrossValidation
from sklearn.model_selection import cross_val_score
cv_score_lr = cross_val_score(LinearRegression(),x,y,cv=5) #cv=no of folds i.e different cunks of data.it automatically
```

```python
print(cv_score_lr)
```

```python
mean_accuracy_lr = sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_lr = mean_accuracy_lr*100
mean_accuracy_lr = round(mean_accuracy_lr,2)
print(mean_accuracy_lr)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
parameters = {'alpha': [0.1, 1.0, 10.0, 100.0]}
ridge_regressor = Ridge()
grid_search = GridSearchCV(ridge_regressor, parameters, cv=5)
grid_search.fit(x_train, y_train)
```

```python
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```
